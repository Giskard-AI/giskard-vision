```python
from time import time
from giskard_vision.landmark_detection.dataloaders.loaders import DataLoader300W
from giskard_vision.landmark_detection.dataloaders.wrappers import CroppedDataLoader, CachedDataLoader
from giskard_vision.landmark_detection.marks.facial_parts import FacialParts
from giskard_vision.landmark_detection.visualisation.draw import draw_marks
```

```python
ds = DataLoader300W(dir_path="../datasets/300W/sample")
```

```python
image, marks, meta = ds[0]
image, marks, meta = image[0], marks[0], meta[0]
draw_marks(image, [marks], ["green"], [True])
```

```python
cropped_ds = CroppedDataLoader(ds, FacialParts.ENTIRE.value)
image, marks, meta = cropped_ds[0]
image, marks, meta = image[0], marks[0], meta[0]
draw_marks(image, [marks], ["green"], [True])
```

```python
cropped_ds = CroppedDataLoader(ds, FacialParts.LEFT_EYE.value)
image, marks, meta = cropped_ds[0]
image, marks, meta = image[0], marks[0], meta[0]
draw_marks(image, [marks], ["green"], [True])
```

```python
ts = time()
dataset = CroppedDataLoader(ds, FacialParts.LEFT_EYE.value)
[elt for elt in dataset]
te = time()
print(round(te - ts, 5), "s.")
```

```python
ts = time()
dataset = CroppedDataLoader(ds, FacialParts.LEFT_EYE.value)
[elt for elt in dataset]
[elt for elt in dataset]
[elt for elt in dataset]
[elt for elt in dataset]
te = time()
print(round(te - ts, 5), "s.")
```

```python
ts = time()
dataset = CachedDataLoader(CroppedDataLoader(ds, FacialParts.LEFT_EYE.value))
[elt for elt in dataset]
[elt for elt in dataset]
[elt for elt in dataset]
[elt for elt in dataset]
te = time()
print(round(te - ts, 5), "s.")
```
