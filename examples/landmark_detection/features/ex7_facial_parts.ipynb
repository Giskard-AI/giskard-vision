```python
from giskard_vision.landmark_detection.marks.facial_parts import FacialParts
from giskard_vision.landmark_detection.dataloaders.loaders import DataLoader300W
from giskard_vision.landmark_detection.visualisation.draw import draw_marks
from giskard_vision.landmark_detection.transformations.crop import crop_mark

ds = DataLoader300W(dir_path="../datasets/300W/sample")
image, marks, meta = ds[4]
image, marks, meta = image[0], marks[0], meta.data
```

```python
facial_part = FacialParts.EYEBROWS.value - FacialParts.LEFT_EYEBROW.value
draw_marks(image, [crop_mark(marks, facial_part)], ["green"], [True])
```

```python
facial_part = FacialParts.EYEBROWS.value + FacialParts.LEFT_EYE.value - FacialParts.RIGHT_HALF.value
draw_marks(image, [crop_mark(marks, facial_part)], ["green"], [True])
```
